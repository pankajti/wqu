## Wiener Process

a process is weinier process if it follows following two properties :

* 1 the change 
$\Delta z$ 
during a small period of time
$\Delta t$ 
is: 

$$ \Delta Z = \epsilon  \sqrt {\Delta t} $$

    where $\epsilon$ has standard normal distribution $\phi$(0,1)

* Values of  $\Delta z$ for any two small interval $\Delta t $ are independent

* $\Delta z  $ has normal distribution with mean =0
* standard deviation of $\Delta z $ is $\sqrt{\Delta t}$
* varience of $\Delta z $ is $\Delta t$



```python
import numpy as np
import matplotlib.pyplot as plt

def wiener_process(T, N,a,b):
    dt = T / N
    t = np.linspace(0.0, T, N)
    dW = np.sqrt(dt) * np.random.normal(0, 1, N)
    W = np.cumsum(dW)
    return t, W

# Parameters
T = 1.0  # Total time
N = 1000  # Number of steps

# Generate Wiener process
t, W = wiener_process(T, N)

# Plot Wiener process
plt.figure(figsize=(8, 6))
plt.plot(t, W, label='Wiener Process')
plt.title('Wiener Process Simulation')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def ito_process(T, N, mu, sigma, S0):
    dt = T / N
    t = np.linspace(0.0, T, N)
    dW = np.sqrt(dt) * np.random.normal(0, 1, N)
    W = np.cumsum(dW)
    drift = (mu - 0.5 * sigma**2) * t
    diffusion = sigma * W
    S = S0 * np.exp(drift + diffusion)
    return t, S

# Parameters
T = 1.0  # Total time
N = 1000  # Number of steps
mu = 0.8  # Drift
sigma = 0.8  # Volatility
S0 = 100.0  # Initial value

# Generate Ito process (Geometric Brownian Motion)
t, S = ito_process(T, N, mu, sigma, S0)

# Plot Ito process
plt.figure(figsize=(8, 6))
plt.plot(t, S, label='Ito Process (Geometric Brownian Motion)')
plt.title('Ito Process Simulation (Geometric Brownian Motion)')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(False)
plt.show()
```
