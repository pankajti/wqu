```python
import numpy as np
import pandas as pd
from pgmpy.estimators import BicScore
from pgmpy.estimators import HillClimbSearch
from pgmpy.estimators import BayesianEstimator
# Generate (discretised) data with dependencies
data = pd.DataFrame(np.random.randint(0, 3, size=(2500, 8)),
columns=list('ABCDEFGH'));
data['A'] += data['B'] + data['C'];
data['H'] = data['G'] - data['A'];
data_train = data[: int(data.shape[0] * 0.75)];
# Learn network structure
hc = HillClimbSearch(data_train, scoring_method=BicScore(data_train));
model = hc.estimate();
# Learn parameters of the network
model.fit(data_train,
estimator=BayesianEstimator, prior_type="BDeu");
# Test the dataset
data_test = data[int(0.75 * data.shape[0]) : data.shape[0]];
data_test.drop('A', axis=1, inplace=True); # Drop variable to be predicted
prediction = model.predict(data_test);
```

```python

import random

def hill_climbing(problem):
  """
  Hill climbing algorithm.

  Args:
    problem: A problem instance.

  Returns:
    A solution to the problem.
  """

  current = problem.initial_state()
  while True:
    neighbors = problem.neighbors(current)
    if not neighbors:
      return current
    best = max(neighbors, key=problem.evaluate)
    if problem.evaluate(best) > problem.evaluate(current):
      current = best

class EightPuzzle(object):
  """
  Eight puzzle problem.

  The goal state is:

    1  2  3
    4  5  6
    7  8  0
  """

  def __init__(self):
    self.initial_state = lambda: [1, 2, 3, 4, 5, 6, 7, 8, 0]
    self.neighbors = lambda state: [self._swap(state, i, i + 1) for i in range(8) if i % 3 != 2]
    self.evaluate = lambda state: sum(1 for i in range(8) if state[i] != i + 1)

  def _swap(self, state, i, j):
    new_state = list(state)
    new_state[i], new_state[j] = new_state[j], new_state[i]
    return new_state

def main():
  problem = EightPuzzle()
  solution = hill_climbing(problem)
  print(solution)

if __name__ == "__main__":
  main()
```
