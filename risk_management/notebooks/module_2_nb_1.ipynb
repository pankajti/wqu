$E_p$(M.$\phi$) = p(b*$B_u$+p*$S_u$

```python
M0 = np.array((B[0], S[0])) 
M0
```

```python
R = M / M0 - 1 
R
```

```python
P = np.array((0.5, 0.5))
```

```python
P
```

```python
np.dot(P, R) 
```

```python
s = 0.55
```

```python
phi = (1-s, s) 
phi
```

```python
mu = np.dot(phi, np.dot(P, R))
mu
```

```python
sigma = s * R[:, 1].std()
sigma
```

```python
values = np.linspace(0, 1, 25)
values
```

```python
R[:1]
```

```python

mu = [np.dot(((1-s), s), np.dot(P, R))
                for s in values]
mu
```

```python
R[:, 1]
```

```python
sigma = [s * R[:, 1].std() for s in values]
sigma
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.plot(values, mu, lw = 3.0, label='$\mu_p$')
plt.plot(values, sigma, '--', lw = 3.0, label='$\sigma_p$')
plt.legend(loc=0)
plt.xlabel('$s$');
```
