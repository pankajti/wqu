### Calculate returns 

```python
num_rec= 5
```

```python
df = pd.DataFrame()
```

```python
import random
```

```python
random.seed(42)
```

```python
df['close'] = [random.randint(0,100) for i in range(num_rec)]
```

```python
df['pred'] = [random.random() for i in range(num_rec)]
```

```python
df['ret'] =df.close.pct_change()
```

```python
df['predicted_positions'] = np.where(df.pred>.5, 1,-1)
```

```python
df['existing_positions'] = df.predicted_positions.shift(1)
```

```python
df['trades'] =  df.predicted_positions-df.existing_positions.fillna(0)
```

```python
df['strategy_return']= df.existing_positions*df.ret
```

```python
df["positions_long_only"]= df["predicted_positions"]
df["positions_long_only"][df["positions_long_only"]==-1]=0
```

```python
df['existing_positions_long_only'] = df.positions_long_only.shift(1)
```

```python
df['strategy_return_long_only']= df.existing_positions_long_only*df.ret
```
