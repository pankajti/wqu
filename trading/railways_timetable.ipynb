```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
nx.single_source_shortest_path(G, 'BBS ')
```

```python
for i in df[['Source Station Code','Destination station Code' ]].to_dict('records'):
    G.add_edge(i ['Source Station Code'],i['Destination station Code'])
```

```python
G.nodes
```

```python
nx.draw_spring(G)
```

```python
pos = nx.spring_layout(G)
```

```python
nx.draw(G.subgraph(list(G.nodes().keys())[:40]), with_labels=True)
```
