```python
import backtrader as bt
import yfinance as yf
import pandas as pd
ticker = 'TCS.NS'
yf_ticker = yf.Ticker(ticker)
hf_tick_hist=yf.download(ticker, start='2022-01-01', end='2022-11-21')#, actions=True)
```

```python
data_feed= bt.feeds.PandasData(dataname =hf_tick_hist.iloc[:,[0,1,2,3,5]])
```

```python
data_feed.datafields
```

```python
class TestStrategy(bt.Strategy):
    params = dict(fast_p = 10, slow_p = 30)
    def __init__(self ):
        self.sma1 = bt.ind.SMA(period = self.params.fast_p)
        self.sma2 = bt.ind.SMA(period = self.params.slow_p)
        self.sma_crs = bt.ind.CrossOver(self.sma1, self.sma2)


    
    def next(self):
        
        print("got something", self.data.close[0],self.data.volume[0], self.sma1[0],self.sma_crs[0],self.params.fast_p)
        
        if not self.position:
            if self.sma_crs>0:
                self.buy()
        elif self.sma_crs<0:
            self.close()
```

```python
cerebro = bt.Cerebro()
```

```python
cerebro.adddata(data_feed)
```

```python
#cerebro.addstrategy(TestStrategy, fast_p=[12,23])
```

```python
cerebro.optstrategy(TestStrategy, fast_p=range(12,23))
```
